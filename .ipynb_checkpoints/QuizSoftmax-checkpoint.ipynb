## Quiz 1
Write a softmax function and plot its values for different scores

```python
"""Softmax."""

scores = [3.0, 1.0, 0.2]

import numpy as np
%matplotlib inline 

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    y = np.exp(x)/np.sum(np.exp(x), axis=0)
    return(y)

print(softmax(scores))

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

## Quiz 2
What happen when you multiply all the scores by 10? How do the probabilities change?

```python
scores_2 = scores*10
plt.plot(x, softmax(scores_2).T, linewidth = 2)
```

The probabilites get closer to either 0 or 1
